## 随机漫步

### 创建RandomWalk 类

```python
#RandomWalk类需要三个属性：一个是存储随机漫步次数的变量，其他两个是列表，分别存储随机漫步经过的每个点的x坐标和y坐标
#只包含两个方法：方法__init___()和fill_walk()，后者计算随机漫步经过的所有点。
```

```python
#为做出随机决策，将所有可能的选择都存储在一个列表中
#在每次决策时都使用模块random中的choice() 来决定使用哪种选择
#将随机漫步包含的默认点数设置为5000
#创建两个用于存储x值和y值的列表，
并让每次漫步都从点(0, 0)出发。

from random import choice

class RandomWalk:
    """一个生成随机漫步数据的类。"""
    def __init__(self, num_points=5000):
        """初始化随机漫步的属性。"""
        self.num_points = num_points
        
        #所有随机漫步都始于(0, 0)。
        self.x_values = [0]
        self.y_values = [0]
```

### 选择方向

```python
### 方法fill_walk()生成漫步包含的点并决定每次漫步的方向

from random import choice

class RandomWalk:
    """一个生成随机漫步数据的类。"""
    def __init__(self, num_points=5000):
        """初始化随机漫步的属性。"""
        self.num_points = num_points
        
        #所有随机漫步都始于(0, 0)。
        self.x_values = [0]
        self.y_values = [0]

    def fill_walk(self):
        """计算随机漫步包含的所有点。"""
        
        # 不断漫步，直到列表达到指定的长度。
        while len(self.x_values) < self.num_points:
            
            # 决定前进方向(向右或者向左；向上或者向下)以及沿这个方向前进的距离。
            x_direction = choice([1, -1])
            x_distance = choice([0, 1, 2, 3, 4])
            x_step = x_direction * x_distance
            y_direction = choice([1, -1])
            y_distance = choice([0, 1, 2, 3, 4])
            y_step = y_direction * y_distance
            
            # 拒绝原地踏步。
            if x_step == 0 and y_step == 0:
                continue
                
            # 计算下一个点的x值和y值。  -1：最后一个值
            x = self.x_values[-1] + x_step
            y = self.y_values[-1] + y_step
            
            self.x_values.append(x)
            self.y_values.append(y)
```

```python
from random import choice

class RandomWalk:
    """一个生成随机漫步数据的类。"""
    def __init__(self, num_points=5000):
        """初始化随机漫步的属性。"""
        self.num_points = num_points
        
        # 所有随机漫步都始于(0, 0)。
        self.x_values = [0]
        self.y_values = [0]

    def walk(self):
        """计算随机漫步包含的所有点。"""
        
        # 不断漫步，直到列表达到指定的长度。
        while len(self.x_values) < self.num_points:
            
            # 修改后的方向和距离选择
            x_direction = choice([1])  # 去掉-1
            x_distance = choice([0, 1, 2, 3, 4]) 
            x_step = x_direction * x_distance
            
            y_direction = choice([1, -1]) 
            y_distance = choice([0, 1, 2, 3, 4, 5, 6, 7, 8])  # 扩大距离选择范围
            y_step = y_direction * y_distance
            
            # 拒绝原地踏步。
            if x_step == 0 and y_step == 0:
                continue
                
            # 计算下一个点的x值和y值。
            x = self.x_values[-1] + x_step
            y = self.y_values[-1] + y_step
            
            self.x_values.append(x)
            self.y_values.append(y)
```

```python
import matplotlib.pyplot as plt
import random
from suiji import RandomWalk  

while True:
    random.seed()  # 重置随机数种子
    rw = RandomWalk(5000)
    rw.walk()
    
    plt.clf()  # 清除之前的图形
    plt.style.use('classic')
    fig, ax = plt.subplots(figsize=(10, 6), dpi=128)
    
    ax.plot(rw.x_values, rw.y_values, linewidth=1)
    ax.plot(0, 0, 'go', markersize=10)  # 起点
    ax.plot(rw.x_values[-1], rw.y_values[-1], 'ro', markersize=10)  # 终点
    
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
    plt.show()
    
    keep_running = input("Make another walk? (y/n): ")
    if keep_running.lower() == 'n':  # 使用 lower() 确保大小写不敏感
        break
```

```python
from random import choice

class RandomWalk:
    """一个生成随机漫步数据的类。"""
    def __init__(self, num_points=5000):
        """初始化随机漫步的属性。"""
        self.num_points = num_points
        
        # 所有随机漫步都始于(0, 0)。
        self.x_values = [0]
        self.y_values = [0]

    def get_step(self):
        """确定每次漫步的距离和方向。"""
        direction = choice([1, -1])
        distance = choice([0, 1, 2, 3, 4])
        return direction * distance

    def walk(self):
        """计算随机漫步包含的所有点。"""
        
        # 不断漫步，直到列表达到指定的长度。
        while len(self.x_values) < self.num_points:
            
            # 获取x和y方向的步长。
            x_step = self.get_step()
            y_step = self.get_step()
            
            # 拒绝原地踏步。
            if x_step == 0 and y_step == 0:
                continue
                
            # 计算下一个点的x值和y值。
            x = self.x_values[-1] + x_step
            y = self.y_values[-1] + y_step
            
            self.x_values.append(x)
            self.y_values.append(y)
```
